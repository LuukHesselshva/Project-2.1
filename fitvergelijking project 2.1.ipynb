```python
import numpy as np
import matplotlib.pyplot as plt
import math

def verschuivingen(n, d, labda, i):
    return(((2*n*d)/labda)*((1/np.cos(i))-1))

i_fit = np.linspace(0, np.deg2rad(10), 1001)
# groene laser heeft een golflengte van 532 nm
labda = 532e-9 #nm
n = 1.5
d = 1e-3 #m

N_berekenen = verschuivingen(n,d,labda,i_fit)


plt.grid()
plt.title("draaing materiaal uitgezet tegenover aantal verschuivingen")
plt.xlabel("i (graden)")
plt.ylabel("N (aantal verschuivingen)")
plt.plot(i_fit*np.rad2deg(np.pi),N_berekenen)
plt.ylim(bottom=0.001)
plt.xlim(left=0.001)
plt.show()
```

```python
import numpy as np

def dOPL(nl, no, i, d, lamvac):
    thetarad = np.deg2rad(i)
    B = np.arcsin((nl*np.sin(thetarad))/no)
    OA = (d/np.cos(B))*no
    AB = (d*np.tan(B))*nl
    dopl = 2*(OA + AB - (d*no))
    return dopl
```
